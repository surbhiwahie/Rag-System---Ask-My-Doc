```python
from Evaluation.golden_dataset import golden_data

def evaluate_system(answer_function):
    
    results = []
    
    for item in golden_data:
        question = item["question"]
        expected = item["ground_truth"]
        
        generated = answer_function(question)
        
        print("\n====================")
        print(f"Q: {question}")
        print(f"Expected: {expected}")
        print(f"Generated: {generated}")
        
        results.append({
            "question": question,
            "expected": expected,
            "generated": generated
        })
    
    return results
```
